```python
import os
from pathlib import Path
os.makedirs("training_data", exist_ok = True)
```

```python
metadata = []
for i in range(1,12):
  metadata.append({"file_name":f"{i:02d}.jpg",
                   "text":f"a photo of ABJ"})
```

```python
print(metadata)
```

```python
import json
with open("training_data/metadata.jsonl","w") as f:
  for item in metadata:
    f.write(json.dumps(item) + "\n")

```
